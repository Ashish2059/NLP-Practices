Merging two pdf files into single file

```python
pdf1 = fitz.open('sample.pdf')
num_pages = pdf1.page_count
print(f'The first PDF file has {num_pages} pages')

pdf2 = fitz.open('sample2.pdf')
num_pages = pdf2.page_count
print(f'The second PDF file has {num_pages} pages')

# create a new PDF object to hold the merged PDFs
merged_pdf = fitz.open()

# loop through the pages of the first PDF file and add them to the new PDF object
for page in pdf1:
    merged_pdf.insert_pdf(pdf1, from_page=page.number, to_page=page.number)
# loop through the pages of the second PDF file and add them to the new PDF object
for page in pdf2:
    merged_pdf.insert_pdf(pdf2, from_page=page.number, to_page=page.number)

# save the merged PDF to a file
merged_pdf.save('merged.pdf')
num_pages = merged_pdf.page_count
print(f'The mergedPDF file has {num_pages} pages')
print(merged_pdf)


# close the PDF files
pdf1.close()
pdf2.close()
merged_pdf.close()
```

#Collecting data from HTML

```python
!pip install bs4
import urllib.request as urllib2
from bs4 import BeautifulSoup
```

```python
response = urllib2.urlopen('https://en.wikipedia.org/wiki/Natural_language_processing')
html_doc = response.read()
```

```python
#Parsing
soup = BeautifulSoup(html_doc, 'html.parser')
# Formating the parsed html file
strhtm = soup.prettify()
# Print few lines
print (strhtm[:1000])
```

```python
#Extracting tag value
print(soup.title)
print(soup.title.string)
print(soup.a.string)
print(soup.b.string)
```

```python
#Extracting all instances of particular tag
#instances of a tag
for x in soup.find_all('a'): 
  print(x.string)
```

```python
#instances of p tag
for x in soup.find_all('p'): 
  print(x.text)
```
